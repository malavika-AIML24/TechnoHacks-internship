<h1 style="color:blue;">K-Means Clustering <h1>

```python
#import Required Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
```

```python
#Load the Dataset
```

```python
iris = datasets.load_iris()
X = iris.data
y = iris.target  
df = pd.DataFrame(X, columns=iris.feature_names)
print(df.head())
```

```python
#Standardize the Data
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
#Applying K-Means Clusterin
```

```python
Kmeans = KMeans(n_clusters=3, random_state=42)
kmeans.fit(X_scaled)
labels = kmeans.labels_
print("Cluster labels:", labels)
```

```python
#Add cluster Label to Data
```

```python
df['Cluster'] = labels
print(df.head())
```

```python
#Visualize the Cluster
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(X_scaled[:, 0], X_scaled[:, 1], c=labels, cmap='viridis', s=50)
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1],  color='red', marker='X', s=200, label='Centroids')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('K-Means Clustering on Iris Dataset')
plt.legend()
plt.show()
```

```python
#Evaluate with inertia
```

```python
print("Inertia (Sum of Squared Distances):", kmeans.inertia_)
```

```python
#3D view Cluster
```

```python
from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure(figsize=(8,6))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(X_scaled[:, 0], X_scaled[:, 1], X_scaled[:, 2],c=kmeans.labels_, cmap='viridis', s=50)
ax.set_title('3D View of Clusters')
ax.set_xlabel('Feature 1')
ax.set_ylabel('Feature 2')
ax.set_zlabel('Feature 3')
plt.show()
```
